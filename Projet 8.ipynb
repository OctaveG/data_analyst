Le dataframe "credit" a moins d'individu car les agrégats de pays n'ont pas été récupérés, pour ne pas fausser les résultats avec une contabalisation multiple de certains pays. 

Ces aggrégats retirés des autres dataframe ultérieurement grâce à une jointure interne gauche. 

```python
population= population.reset_index().melt(id_vars='country_name')

population.columns = ['country_name','year','population']

population= population.set_index(['country_name','year'], drop=True)

population= population.sort_index()

population.head()
```

```python
pib= pib.reset_index().melt(id_vars='country_name')

pib.columns = ['country_name','year','pib']

pib= pib.set_index(['country_name','year'], drop=True)

pib= pib.sort_index()

pib.head()
```

```python
credit= credit.reset_index().melt(id_vars='country_name')

credit.columns = ['country_name','year','private_domestic_credit_percentage_GDP']

credit= credit.set_index(['country_name','year'], drop=True)

credit= credit.sort_index()

credit.head()
```

```python
croissance = croissance.reset_index().melt(id_vars='country_name')

croissance.columns = ['country_name','year','growth_percentage_gdp']

croissance = croissance.set_index(['country_name','year'], drop=True)

croissance = croissance.sort_index()

croissance.head()
```

```python
balance = balance.reset_index().melt(id_vars='country_name')

balance.columns = ['country_name','year','current_account_balance_percentage_gdp']

balance = balance.set_index(['country_name','year'], drop=True)

balance = balance.sort_index()

balance.head()
```

```python
inflation = inflation.reset_index().melt(id_vars='country_name')

inflation.columns = ['country_name','year','inflation']

inflation = inflation.set_index(['country_name','year'], drop=True)

inflation = inflation.sort_index()

inflation.head()
```

```python
chomage = chomage.reset_index().melt(id_vars='country_name')

chomage.columns = ['country_name','year','unemployment_pourcentage_total_labor_force']

chomage = chomage.set_index(['country_name','year'], drop=True)

chomage = chomage.sort_index()

chomage.head()
```

```python
tbd_kaldor = population.merge(pib, how ='inner', on=['country_name','year'])

tbd_kaldor = tbd_kaldor.merge(credit, how ='inner', on=['country_name','year'])

tbd_kaldor = tbd_kaldor.merge(chomage, how ='inner', on=['country_name','year'])

tbd_kaldor = tbd_kaldor.merge(croissance, how ='inner', on=['country_name','year'])

tbd_kaldor = tbd_kaldor.merge(inflation, how ='inner', on=['country_name','year'])

tbd_kaldor = tbd_kaldor.merge(balance, how ='inner', on=['country_name','year'])

tbd_kaldor.head()
```

```python
tbd_kaldor
```

```python
tbd_kaldor.unemployment_pourcentage_total_labor_force = tbd_kaldor.unemployment_pourcentage_total_labor_force * tbd_kaldor.population / 100            



tbd_kaldor.private_domestic_credit_percentage_GDP = tbd_kaldor.private_domestic_credit_percentage_GDP * tbd_kaldor.pib / 100

tbd_kaldor.growth_percentage_gdp = tbd_kaldor.growth_percentage_gdp * tbd_kaldor.pib /100

tbd_kaldor.current_account_balance_percentage_gdp = tbd_kaldor.current_account_balance_percentage_gdp * tbd_kaldor.pib / 100
```

```python
tbd_kaldor.columns = ['population', 'gdp_current_dollar', 'private_domestic_credit_current_dollar',
       'unemployment', 'growth_current_dollar',
       'inflation', 'current_account_balance_current_dollar']
```
