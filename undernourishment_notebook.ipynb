On défini une clé primaire parmi les clés candidates de nos tables, à l'aide d'une fonction. 
On retient comme clé candidate : ['pays', 'produit','unite']

```python
def primary_key(dataframe,cle_candidate):
    if dataframe.dropna(subset=cle_candidate).drop_duplicates(subset=cle_candidate, keep=False).shape == dataframe.shape: 
        return print('This is a Primary key') 
    else: 
        return print('This might help you: https://en.wikipedia.org/wiki/Primary_key')
```

```python
# importation des tables 

population = pd.read_csv('/Users/octav/Google Drive/OC_Data Analyst_GORAM Octave/Projets/Projet 3/DF_FAO/Questions 15 et suivantes/population.csv')

animaux = pd.read_csv('/Users/octav/Google Drive/OC_Data Analyst_GORAM Octave/Projets/Projet 3/DF_FAO/Questions 15 et suivantes/animaux.csv')

vegetaux = pd.read_csv('/Users/octav/Google Drive/OC_Data Analyst_GORAM Octave/Projets/Projet 3/DF_FAO/Questions 15 et suivantes/vegetaux.csv')

equilibre_prod = pd.read_csv('/Users/octav/Google Drive/OC_Data Analyst_GORAM Octave/Projets/Projet 3/DF_FAO/Questions 15 et suivantes/equilibre_prod.csv')

sous_nutrition = pd.read_csv('/Users/octav/Google Drive/OC_Data Analyst_GORAM Octave/Projets/Projet 3/DF_FAO/Questions 15 et suivantes/sous_nutrition.csv')
```

```python
# formatage table population

population = population[['Pays','Code Pays','Année','Valeur']]

population.columns = ["pays", "code_pays", "annee", "population"]

population['population'] = population['population'] * 1000
```

```python
# formatage table dispo_alim

animaux['origine'] = 'animale'

vegetaux['origine'] = 'vegetale'

dispo_alim = animaux.append(vegetaux)

dispo_alim.columns = ['Code Domaine', 'Domaine', 'code_pays', 'pays', 'Code Élément',
       'elément', 'code_produit', 'produit', 'Code Année', 'annee', 'Unité',
       'valeur', 'Symbole', 'Description du Symbole', 'origine']

dispo_alim = dispo_alim.pivot_table(
index=["pays", "code_pays", "annee", "produit", "code_produit", "origine"],
    columns = ["elément"], values=["valeur"], aggfunc=sum)

dispo_alim.columns = ['dispo_alim_kcal_p_j', 'dispo_alim_tonnes', 'dispo_mat_gr', 'dispo_prot']

dispo_alim = dispo_alim.reset_index()

dispo_alim['dispo_alim_tonnes'] = dispo_alim['dispo_alim_tonnes'] / 1000

del animaux, vegetaux
```

```python
# formatage table equilibre_prod

equilibre_prod.columns = ['Code Domaine', 'Domaine', 'code_pays', 'pays', 'Code Élément',
       'elément', 'code_produit', 'produit', 'Code Année', 'annee', 'Unité',
       'valeur', 'Symbole', 'Description du Symbole']

equilibre_prod = equilibre_prod.pivot_table(
index=["pays", "code_pays", "annee", "produit", "code_produit"],
    columns = ["elément"], values=["valeur"], aggfunc=sum)

equilibre_prod.columns = ['alim_ani','autres_utilisations', 'dispo_int', 'nourriture', 'pertes', 'semences',  'transfo' ]


equilibre_prod = equilibre_prod.reset_index()
```

```python
# formatage table sous_nutrition

sous_nutrition = sous_nutrition[['Zone','Code zone','Année','Valeur']]

sous_nutrition.columns = ["pays", "code_pays", "annee", "nb_personnes"]

sous_nutrition = sous_nutrition.loc[sous_nutrition.nb_personnes.str.contains("<") == False]

sous_nutrition[['nb_personnes']] = sous_nutrition[['nb_personnes']].astype(float)

sous_nutrition[['nb_personnes']] = sous_nutrition[['nb_personnes']] * 1000000
```

```python
# supression des pays doublons
population = population.drop(population[population.pays.isin(["Chine - RAS de Hong-Kong", "Chine - RAS de Macao","Chine, continentale","Chine, Taiwan Province de"])].index)

dispo_alim = dispo_alim.drop(dispo_alim[dispo_alim.pays.isin(["Chine - RAS de Hong-Kong", "Chine - RAS de Macao","Chine, continentale","Chine, Taiwan Province de"])].index)

equilibre_prod = equilibre_prod.drop(equilibre_prod[equilibre_prod.pays.isin(["Chine - RAS de Hong-Kong", "Chine - RAS de Macao","Chine, continentale","Chine, Taiwan Province de"])].index)

sous_nutrition = sous_nutrition.drop(sous_nutrition[sous_nutrition.pays.isin(["Chine - RAS de Hong-Kong", "Chine - RAS de Macao","Chine, continentale","Chine, Taiwan Province de"])].index)
```

```python
# reset index
dispo_alim = dispo_alim.reset_index()

equilibre_prod = equilibre_prod.reset_index()

population = population.reset_index()

sous_nutrition = sous_nutrition.reset_index()
```
